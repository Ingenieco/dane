```python
emicron22 = pd.read_csv('datos/emicron_22.csv')
emicron22 = emicron22.rename(columns = lambda x: x.lower())
emicron22.head()
```

```python
df_resultado = pd.merge(merge_df, emicron22, left_on='directorio_hbos', right_on='directorio', how = 'inner')
```

```python
df_resultado.head()
```

```python
nuevas_columnas = {col: col.replace('_y', '') for col in df_resultado.columns}
df_resultado.rename(columns = nuevas_columnas, inplace = True)
```

```python
emicron_anomalias = df_resultado[emicron22.columns]
emicron_anomalias.head()
```

```python
emicron_anomalias.shape
```
